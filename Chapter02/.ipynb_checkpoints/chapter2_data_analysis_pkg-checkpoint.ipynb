## Scipy

```python
from scipy.optimize import fsolve
```

```python
def f(x):
    x1 = x[0]
    x2 = x[1]
    return [2*x1 - x2**2 - 1, x1**2 - x2 - 2]
```

```python
result = fsolve(f, [1,1])
```

```python
print(result, [int(i) for i in f(result)])
```

```python
from scipy import integrate
```

```python
def g(x):
    return (1-x**2)**0.5
```

```python
pi_2, err = integrate.quad(g, -1, 1)
```

```python
print(pi_2*2)
```

## Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 1000)
y = np.sin(x) + 1
z = np.cos(x**2) + 1

plt.figure(figsize=(8,4))
plt.plot(x, y, label='$\sin x+1$', color='red', linewidth=2)
plt.plot(x, z, 'b--', label='$\cos x**2+1$')
plt.xlabel('Time(s)')
plt.ylabel('Volt')
plt.title('A Simple Excample',fontdict={'fontsize':20})
plt.ylim(0, 2.2)
plt.legend()
plt.show()
```
