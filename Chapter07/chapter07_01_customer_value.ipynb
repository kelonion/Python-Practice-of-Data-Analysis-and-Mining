```python
import pandas as pd
```

```python
datafile = './data/air_data.csv'
```

```python
data = pd.read_csv(datafile, encoding='utf-8')
```

```python
data.head().T
```

```python
explore = data.describe(percentiles=[], include='all').T
```

```python
explore['null'] = len(data) - explore['count']
```

```python
explore
```

```python
data_sum_na = data[data['SUM_YR_1'].isna() | data['SUM_YR_2'].isna()]
```

```python
data_sum_na.tail().T
```

```python
data_c = data[data['SUM_YR_1'].notnull() & data['SUM_YR_2'].notnull()]
```

```python
cond1 = data_c['SUM_YR_1'] != 0
cond2 = data_c['SUM_YR_2'] != 0
cond3 = (data_c['SEG_KM_SUM'] == 0) &(data_c['avg_discount'] == 0)
```

```python
data_c = data_c[cond1 | cond2 | cond3]
```

```python
LRFMC = ['LOAD_TIME', 'FFP_DATE', 'LAST_TO_END', 'FLIGHT_COUNT', 'SEG_KM_SUM', 'avg_discount']
data_to_als = data_c[LRFMC]
```

```python
data_to_als['LOAD_TIME'] = pd.to_datetime(data_to_als['LOAD_TIME']) - pd.to_datetime(data_to_als['FFP_DATE'])
```

```python
data_to_als = data_to_als.drop(columns=['FFP_DATE'])
```

```python
data_to_als.columns = ['L', 'R', 'F', 'M', 'C']
```

```python
data_to_als.head()
```

```python
data_to_als.info()
```

```python
data_cn = (data_to_als - data_to_als.mean(axis=0)) / data_to_als.std(axis=0)
```

```python
from sklearn.cluster import KMeans
```

```python
k = 5
```

```python
kmodel = KMeans(n_clusters=k, n_jobs=-1)
kmodel.fit(data_cn)
```

```python
data_cn['cluster'] = kmodel.predict(data_cn)
```

```python
data_cn['cluster'].value_counts()
```

```python
data_c['cluster'] = data_cn['cluster']
```

```python
data_c['SUM'] = data_c['SUM_YR_1'] + data_c['SUM_YR_2']
```

```python
col = list(data_c.columns)
col = [col[-1]] + col[:-1]
data_c = data_c[col]
```

```python
data_cls = data_c.groupby(['cluster']).mean().sort_values(by=['SUM'], ascending=False)
```
