```python
import pandas as pd
```

```python
datafile = './data/air_data.csv'
data = pd.read_csv(datafile, encoding='utf-8')
```

```python
data.head().T
```

```python
data_6 = data[data['FLIGHT_COUNT'] > 6]
```

```python
data_6['FLIGHT_RATIO'] = data['L1Y_Flight_Count'] / data['P1Y_Flight_Count']
```

```python
data_6.loc[data_6['FLIGHT_RATIO'] < 0.5,'CUSTOMER_TYPE'] = 0
data_6.loc[(data_6['FLIGHT_RATIO'] >= 0.5) & (data_6['FLIGHT_RATIO'] <0.9), 'CUSTOMER_TYPE'] = 1
data_6.loc[data_6['FLIGHT_RATIO'] >= 0.9,'CUSTOMER_TYPE'] = 2
```
